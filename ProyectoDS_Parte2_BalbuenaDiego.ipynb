# Proyecto Calidad del Vino

La base de datos es extraida de Kaggle, muestra distintas variables que determinan la calidad del vino.

# Descripción de las variables
 Acidez Fija: La acidez fija se refiere a la concentración de ácidos no volátiles presentes en el vino, principalmente ácido tartárico, ácido málico y ácido cítrico. Estos ácidos son importantes para la percepción del sabor y la estabilidad del vino.

Acidez Volátil: La acidez volátil se refiere a la concentración de ácidos volátiles presentes en el vino, principalmente ácido acético. Niveles altos de acidez volátil pueden indicar una fermentación defectuosa o una contaminación bacteriana, lo que puede afectar negativamente el sabor y el aroma del vino.

Ácido Cítrico: El ácido cítrico es un ácido orgánico presente naturalmente en muchas frutas cítricas. En el vino, el ácido cítrico puede contribuir a la acidez total y afectar el equilibrio de sabores y la frescura del vino.

Azúcar Residual: El azúcar residual se refiere a la cantidad de azúcar que queda en el vino después del proceso de fermentación. Los vinos con niveles más altos de azúcar residual tienden a ser más dulces, mientras que los vinos con niveles bajos o nulos de azúcar residual se consideran secos.

Cloruros: Los cloruros en el vino provienen principalmente de la adición de ácido clorhídrico durante la vinificación o de la utilización de productos de limpieza y desinfección. Altos niveles de cloruros pueden indicar contaminación o procesos de elaboración inadecuados.

Dióxido de Azufre Libre: El dióxido de azufre libre es un compuesto que se utiliza comúnmente en la vinificación como antioxidante y agente antimicrobiano. Puede contribuir a la estabilidad y longevidad del vino al prevenir la oxidación y el crecimiento microbiano no deseado.

Dióxido de Azufre Total: El dióxido de azufre total es la suma del dióxido de azufre libre y el dióxido de azufre unido a otros compuestos en el vino. También juega un papel importante en la preservación del vino y en la prevención de la oxidación y la contaminación microbiana.

Densidad: La densidad del vino se refiere a la masa por unidad de volumen y está influenciada por la concentración de azúcares y alcohol. Puede ser un indicador de la madurez del vino y está relacionada con su cuerpo y textura.

pH: El pH es una medida de la acidez o alcalinidad del vino. Los vinos con un pH más bajo tienden a ser más ácidos, mientras que los vinos con un pH más alto tienden a ser más alcalinos. El pH puede afectar la estabilidad del vino, así como su sabor y estructura.

Sulfatos: Los sulfatos son compuestos que se añaden a veces al vino como conservantes antioxidantes. Ayudan a prevenir la oxidación y el deterioro del vino, contribuyendo así a su longevidad.

Alcohol: El contenido de alcohol en el vino se refiere a la cantidad de alcohol etílico presente, expresado típicamente como un porcentaje de volumen. El nivel de alcohol puede influir en el sabor, el cuerpo y la sensación en boca del vino, así como en su capacidad de envejecimiento.

calidad: representa numéricamente la calidad del vino.

color: clasifica segun el vino sea blanco o rojo.

# Analisis distributivo de las variables

```python
data_ph = df[['pH','color']].groupby(by='color').count()
data_ph
```

```python
data_vola = df[['acidez volatil','color']].groupby(by='color').mean()
data_vola
```

```python
data_vola.rename(columns={'acidez volatil': 'acidez_volatil'}, inplace=True)
data_vola
```

```python
ph_torta = df[['pH','color']].groupby(by='color').mean()
ph_torta
```

```python
plt.figure(figsize=(9,4))
plt.subplot(221)
plt.pie(data_ph.pH, labels = data_ph.index,autopct='%1.1f%%')
plt.title('Distribucion segun color del vino')
plt.subplot(222)
plt.pie(data_vola.acidez_volatil, labels = data_vola.index,autopct='%1.1f%%')
plt.title('Proporciones de acidez volatil')
plt.subplot(223)
plt.pie(ph_torta.pH, labels = ph_torta.index,autopct='%1.1f%%')
plt.title('Equilibrio del PH')
plt.subplot(224)
plt.pie(data_alcohol.alcohol, labels = data_alcohol.index,autopct='%1.1f%%')
plt.title('Graduacion de Alcohol')
```

# Visualización de outliers

```python
fig, axes = plt.subplots(4, 4, figsize=(12, 8))
fig.suptitle('Boxplots de Variables', fontsize=12)

axes = axes.flatten()
for i, column in enumerate(df.columns):

    if column != 'color':

      ax = axes[i]

      ax.boxplot(df[column])
      ax.set_xlabel(column, fontsize = 8)
      ax.set_ylabel('Frecuencia')
      ax.set_xticks([])  # Elimine los ticks en el eje x (porque aparecía el índice "1" que indica la posición en este arreglo unidimensional)

plt.tight_layout()
```

```python
# Visualizar la distribución de la calidad del vino
sns.histplot(df['calidad'], bins=10, kde=True)
plt.xlabel('Calidad del vino')
plt.ylabel('Frecuencia')
plt.title('Distribución de la calidad del vino')
plt.show()
```

```python
count = df['calidad'].value_counts().sort_index().reset_index()

count
```

Como puede observarse tanto en el histograma/boxplot, como en la tabla de distribución de frecuencias, hay más vinos de calidad 5 y 6.
Los outliers indican que hay vinos de calidades muy diferentes en este conjunto de datos. Los outliers hacen referencia a vinos muy buenos o muy malos.

# Gráficos de dispersión

```python
fig, axes = plt.subplots(4, 4, figsize=(12, 8))
fig.suptitle('gráficos de dispersión por calidad', fontsize=12)

axes = axes.flatten()

for i, column in enumerate(df.columns):
    ax = axes[i]
    ax.scatter(df.calidad, df[column], color= 'blue', s=4) #s= ajusta el tamaño de los puntos del gráfico.
    ax.set_xlabel('Calidad', fontsize = 8)
    ax.set_ylabel(column)

plt.tight_layout()
```

No se visualiza un patrón claro que indique una tendencia acerca del nivel de cada propiedad química que un vino debiera tener para ser de determinada calidad.
Esto puede a que el comportamiento de cada elemento no es independiente en la determinacion de la calidad.
